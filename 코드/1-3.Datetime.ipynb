##3.데이터 전처리

```python
#NaN 개수 확인

df.isna().sum()
```

```python
#NaN 행 삭제

df = df.dropna()
df.isna().sum()
```

```python
#행렬 전환

df = df.set_index('지역')
df = df.transpose()
df.head()
```

```python
#datetime으로 변환
#날짜를 index로

df = df.reset_index()
df['index'] = pd.to_datetime(df['index'], format='%y.%b')
df = df.rename({'index' : '날짜'}, axis=1)
df = df.set_index('날짜')
df.head()
```

```python
#dytpe int로 변환

df = df.astype(int)
df.head()
```
